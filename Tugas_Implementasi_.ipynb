```python
import numpy as np
import time
import matplotlib.pyplot as plt

def f(x):
    return 4 / (1 + x**2)

def metode_trapesium(a, b, N):
    x = np.linspace(a, b, N+1)
    y = f(x)
    h = (b - a) / N
    integral = (h/2) * np.sum(y[:-1] + y[1:])
    return integral

def hitung_rms(errors):
    return np.sqrt(np.mean(np.square(errors)))

# Pengujian implementasi
a = 0
b = 1
N_values = [10, 100, 1000, 10000]
pi_referensi = 3.14159265358979323846

errors = []
times = []

for N in N_values:
    start_time = time.time()
    pi_approx = metode_trapesium(a, b, N)
    elapsed_time = time.time() - start_time
    error = np.abs(pi_referensi - pi_approx)
    errors.append(error)
    times.append(elapsed_time)
    print(f"N = {N}, Aproksimasi Pi = {pi_approx}, Error = {error}, Waktu = {elapsed_time:.5f} detik")

rms_error = hitung_rms(errors)
print(f"Galat RMS: {rms_error}")

# Plotting galat dan waktu eksekusi
plt.figure(figsize=(12, 5))

# Plot Galat
plt.subplot(1, 2, 1)
plt.plot(N_values, errors, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Jumlah Segmen (N)')
plt.ylabel('Galat')
plt.title('Galat Aproksimasi vs Jumlah Segmen')
plt.grid(True)

# Plot Waktu Eksekusi
plt.subplot(1, 2, 2)
plt.plot(N_values, times, marker='o')
plt.xscale('log')
plt.xlabel('Jumlah Segmen (N)')
plt.ylabel('Waktu Eksekusi (detik)')
plt.title('Waktu Eksekusi vs Jumlah Segmen')
plt.grid(True)

plt.tight_layout()
plt.show()
```
